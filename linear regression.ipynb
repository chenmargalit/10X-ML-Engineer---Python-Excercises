```python
import numpy as np
from scipy.stats import linregress

def is_linear(data_x, data_y, threshold=0.05):
    """
    Determine if the given data appears to follow a linear pattern.

    Parameters:
    - data_x: array-like, x-values of the data
    - data_y: array-like, y-values of the data
    - threshold: float, p-value threshold for hypothesis testing

    Returns:
    - result: bool, True if the data appears linear, False otherwise
    """

    # Perform linear regression
    slope, intercept, r_value, p_value, _ = linregress(data_x, data_y)

    # Check if the p-value is below the threshold
    return p_value < threshold

# Example usage:
x_values = np.array([1, 2, 3, 4, 5])
y_values = np.array([2, 4, 5, 4, 5])

result = is_linear(x_values, y_values)
print(f"Is the data linear? {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple dataset
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting using different types of plots
plt.figure(figsize=(15, 8))

# Plot 1: Line plot
plt.subplot(2, 3, 1)
plt.plot(x, y, label='Line Plot')
plt.title('Line Plot')

# Plot 2: Scatter plot
plt.subplot(2, 3, 2)
plt.scatter(x, y, label='Scatter Plot', color='orange')
plt.title('Scatter Plot')

# Plot 3: Bar plot
plt.subplot(2, 3, 3)
plt.bar(x, y, label='Bar Plot', color='green')
plt.title('Bar Plot')

# Plot 4: Histogram
plt.subplot(2, 3, 4)
plt.hist(y, bins=20, label='Histogram', color='purple', alpha=0.7)
plt.title('Histogram')

# Plot 5: Pie chart
plt.subplot(2, 3, 5)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```
