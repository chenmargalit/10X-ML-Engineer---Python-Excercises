# Limits

A limit is the value that a function approaches as the input <b> approaches </b> a certain value.

## Intuitive Understanding of Limits

A limit is the value that a function approaches as the input (or variable) approaches a certain value. Limits are fundamental in calculus and are used to define derivatives and integrals.

- **Example**: Consider the function f(x) = x^2. As x approaches 2, f(x) approaches 4. We write this as:
  $$
  \lim_{{x \to 2}} f(x) = 4
  $$


```python
def f(x):
    return x**2

x = np.linspace(0, 4, 400)
y = f(x)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.lineplot(x=x, y=y, label=r'$f(x) = x^2$')

plt.axvline(x=2, color='red', linestyle='--', label=r'$x \to 2$')

plt.text(2, 5, r'$\lim_{{x \to 2}} f(x) = 4$', color='red', ha='left')

plt.xlabel('x')
plt.ylabel(r'$f(x)$')
plt.title(r'Plot of $f(x) = x^2$ with the Limit as $x \to 2$')
plt.legend()

plt.show()
```

## One-Sided Limit

- **Left-hand limit**: The value the function approaches as x approaches a from the left. $$x \to a^-$$

  $$
  \lim_{{x \to a^-}} f(x)
  $$
- **Right-hand limit**: The value the function approaches as as x approaches a from the right. $$x \to a^+$$
  $$
  \lim_{{x \to a^+}} f(x)
  $$

```python
def f(x):
    return x**2

x = np.linspace(0, 4, 400)
y = f(x)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.lineplot(x=x, y=y, label=r'$f(x) = x^2$')

plt.axvline(x=2, color='red', linestyle='--', label=r'$x \to 2$')

plt.text(2, 5, r'$\lim_{{x \to 2}} f(x) = 4$', color='red', ha='left')

plt.xlabel('x')
plt.ylabel(r'$f(x)$')
plt.title(r'Plot of $f(x) = x^2$ with the Limit as $x \to 2$')
plt.legend()

plt.show()
```

## Piecewise function

A Piecewise function is a function where different rules are applied to different parts of the function

```python
class PiecewiseFunction:
    
    def evaluate(self, x):
        
        if x == 4:
            return 5
        else:
            return x ** 2 + x
```

```python
piecewise_function = PiecewiseFunction()

x_values = np.linspace(-8, 8)
y_values = np.array([piecewise_function.evaluate(x) for x in x_values])

x_special = 4
y_special = piecewise_function.evaluate(x_special)

sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))

sns.lineplot(x=x_values, y=y_values, label="f(x) = x^2 + x for x != 4")
plt.scatter(x_special, y_special, color='red', label="f(4) = 5")

plt.title("Piecewise Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

## Techniques for Finding Limits

**1. Direct Substitution**

If \( f(x) \) is continuous at \( x = a \), then:
$$
\lim_{{x \to a}} f(x) = f(a)
$$

**2. Factoring**

Factor the function and simplify to find the limit.
- **Example**:
  $$
  \lim_{{x \to 2}} \frac{x^2 - 4}{x - 2}
  $$
  Factor the numerator:
  $$
  \lim_{{x \to 2}} \frac{(x - 2)(x + 2)}{x - 2} = \lim_{{x \to 2}} (x + 2) = 4
  $$

## Inflection point

Nekudat pitul

An inflection point on a curve is a point where the curve changes concavity, meaning it transitions from being concave upwards (where the curve is bending upwards like a smiley face) to concave downwards (where the curve is bending downwards like a frown), or vice versa.

### Second derivative

The second derivative of a function is the derivative of the derivative of that function. It provides information about the curvature or concavity of the function.

```python
x = sp.Symbol('x')
f = x**3 - 3*x

f_double_prime = sp.diff(f, x, 2)

inflection_points = sp.solve(f_double_prime, x)

inflection_points = [float(point.evalf()) for point in inflection_points]
```

```python
x_vals = np.linspace(-2, 2, 400)
y_vals = [f.subs(x, val) for val in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 3x')

for point in inflection_points:
    plt.scatter(point, f.subs(x, point), color='red', label='Inflection Point')

plt.title('Function with Inflection Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

print("Inflection point(s) found at x =", inflection_points)
```

## The fundamental theorem of calculus

The fundamental theorem of calculus defines a relationship between differentiation and integration. In simple words it says that if I define a function F(x) that is the area between some function and two borders (right border is x), then its derivative is going to be the function itself. This rule explains the connection between a graph (F) and the function that holds it (f). This connection, this transformation, is a derivative.

It has 2 parts:

Part 1 (also known as the First Fundamental Theorem of Calculus):

This part states that if we have a continuous function f(x) on an interval [a, b], and we define a new function F(x) as the definite integral of f(t) from a to x, then F(x) is an antiderivative of f(x).

In other words, the derivative of the integral function F(x) is the original function f(x).

$$ F(x) = \int_a^x f(t)\, dt $$ 

Part 2 states that if f(x) is continuous on [a,b], then F(b) - F(a)

$$  \int_a^b f(x) \, dx = F(b) - F(a) $$

The simplest way to compute the distance under a graph is using the FTC rule which states that we can take the function's antiderivative, place the borders in the function, substract the results and take the absolute value of the substraction result.

```python
x, y = sp.symbols('x y')

f = x**3 + y**2

partial_f_x = sp.diff(f, x)
partial_f_y = sp.diff(f, y)

print("Function: f(x, y) = x^2 + y^2")
print("Partial derivative with respect to x: ∂f/∂x =", partial_f_x)
print("Partial derivative with respect to y: ∂f/∂y =", partial_f_y)
```

```python
def f(x, y):
    return x**2 + y**2

x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
plt.title('Function f(x, y) = x^2 + y^2')
ax.dist=7
plt.show()
```

```python
x, y = sp.symbols('x y')

f = x**2 * y + 3*x * y**2

df_dx = sp.diff(f, x)  
df_dy = sp.diff(f, y)  

x0, y0 = 1, 2
df_dx_at_point = df_dx.subs({x: x0, y: y0})
df_dy_at_point = df_dy.subs({x: x0, y: y0})

print(f"Partial derivative with respect to x: {df_dx}")
print(f"Partial derivative with respect to y: {df_dy}")
print(f"Partial derivative with respect to x at ({x0}, {y0}): {df_dx_at_point}")
print(f"Partial derivative with respect to y at ({x0}, {y0}): {df_dy_at_point}")
```

```python
epsilon = 0.001
```

```python
def f_partial(x, y):
    return x**2 * y + 3*x * y**2
```

```python
x,y = 1,2
df_dx = float(df_dx_at_point.evalf())
```

```python
f_partial(x + epsilon, y)
```

```python
f_partial(x, y)
```

```python
df_dx * epsilon
```

```python
np.allclose(f_partial(x + epsilon, y) - f_partial(x, y), df_dx * epsilon, atol=0.0001)
```

## Composite function

```python
def f(x):
    return x ** 2

def g(x):
    return x + 1

def composite(x):
    return f(g(x))

result = composite(3)
print("Result of composite function:", result)  # Output: Result of composite function: 16
```

## Chain rule

```python
def g(x):
    return x**2 + 1

def f(u):
    return np.sqrt(u)

def g_prime(x):
    return 2 * x

def f_prime(u):
    return 1 / (2 * np.sqrt(u))

def h(x):
    return f(g(x))

def h_prime(x):
    return f_prime(g(x)) * g_prime(x)

x = 3.0
print("g(x) =", g(x))                   
print("f(g(x)) =", f(g(x)))             
print("h(x) =", h(x))                   
print("h'(x) =", h_prime(x)) 
```
