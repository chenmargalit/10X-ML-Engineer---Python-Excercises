# Regression vs Classification

Regression and classification are 2 types of supervised learning (e.g. learning when you have the label), but they are quite different.

Regression is used when the target variable (what we want to predict) is continuous. The model output will be continuous. Some number, like 1000 or 3.57.

We would use regression for example predicting house/stock prices, life expectancy, etc.. A good evaluation metric for such a task might be MSE, or R-squared.

Classification is used when we want to predict a categorical variable. Did he or didnt he pass the test, is our next customer going to be younger or older than 18, etc. The output of the model is a discrete value, for example this email is spam, or its not spam.

Appropriate evaluations metrics for that might be Accuracy, Precision, Recall, etc.

The KNN neighbors model can be used for both classification and regression, depending on its implementation.

# Distance Metrics

## Euclidean Distance

Measures the straight-line distance between two points in Euclidean space (basically the geometry we all know and "love")

$$
\text{d}(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$


```python
def euclidean_distance(x1, y1, x2, y2):
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

Used in places where we calculate geometric distance, for example KNN neighbors

## Cosine Similarity

$$
\text{sim}(x, y) = \frac{x \cdot y}{\|x\| \|y\|}
$$

```python
def cosine_similarity(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))
```

Measures the similarity between two vectors based on their orientation (angle) rather than magnitude.

Used for evaluating similarity between documents or terms in text. Used in topic modeling.

## Manhattan Distance (L1 Norm)

$$
d_{\text{Manhattan}}(A, B) = \sum_{i=1}^{n} |A_i - B_i|
$$


```python
A = (2, 1)
B = (4, 3)

distance = sum(abs(a - b) for a, b in zip(A, B))
```

```python
def manhattan_distance(x, y):
    return sum(abs(a - b) for a, b in zip(A, B))
```

Useful for measuring distances between high-dimensional data points or when features are not of the same scale. 

Used in KNN where the features are on differences scales.

## Jaccard Distance

$$
\text{Jaccard Index} = \frac{|A \cap B|}{|A \cup B|}
$$

$$
\text{Jaccard Distance} = 1 - \frac{|A \cap B|}{|A \cup B|}
$$

groceries_list_1 = milk, powder, protein, tomatos
groceries_list_2 = tea, eggplants, protein, tomatos

$$
\text{Jaccard Index} = \frac{overlapping}{total}
$$

```python
def jaccard_distance(set1, set2):
    
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return 1 - intersection / union
```

Used in all sorts of sets comparisons. These sets might be text, purchased items, movies seen, etc.

# K Nearest Nieghbors From Scratch

```python
class KNNNeighbors:

    def __init__(self, k=3, task_type='classification'):
        self.k = k
        self.task_type = task_type  # 'classification' or 'regression'
        
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train  
        
    def predict(self, X_test):
        predictions = [self._predict(x) for x in X_test]
        return np.array(predictions)
    
    def _predict(self, x):
        
        distances = [self._euc_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        if self.task_type == 'classification':
            most_common = Counter(k_nearest_labels).most_common(1)
            return most_common[0][0]
        
        elif self.task_type == 'regression':
            return np.mean(k_nearest_labels)
        
    def _euc_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
k = KNNNeighbors()
```

```python
k.fit(X_train, y_train)
```

```python
y_pred = k.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

# KNN Limitations

1. Curse of dimensionality - the higher the number of dimensions is, the more meaningless the distance between points become. In a 2d space it makes a lot of sense, what does euclidean distance look like in 100d space?. The more dimensions, the exponentially more data we need.
2. Imbalanced data - if there are very few items of some class, and KNN takes a majority vote, it might be difficult to find a majority that fits the rare class.
3. Time complexity of O(nd) n being the number of samples and d the number of features. Its not fast.
