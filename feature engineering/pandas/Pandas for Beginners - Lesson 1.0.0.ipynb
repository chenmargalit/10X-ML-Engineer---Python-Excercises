```python
df.loc[0] # row by label
```

Looks like index and label are the same. By default, they are. The row's label, is a numeric index. You can change this to make your data easier to handle.

```python
rows_labels = []
for i in range(len(df)):
    rows_labels.append(f'row{i}')
```

```python
df.index = rows_labels
```

# Data Cleaning & Processing

```python
data = {
    'Team': ['Spain', None, 'Israel'],
    'Chance_of_winning': [100, None, 0.01],
    'City': ['Barcelona', 'Manchester', None]
}

temp_df = pd.DataFrame(data)
```

As Chance_of_winning is a column of floats, None is "NaN"

```python
temp_df.head()
```

```python
print(temp_df.isnull())
```

```python
df_cleaned = temp_df.dropna()
print(df_cleaned)
```

```python
df_filled = temp_df.fillna(value={'Chance_of_winning': temp_df['Chance_of_winning'].mean()})
print(df_filled)
```

# Concats & Merges

```python
data = {
    'Team': ['Spain', 'England', 'Israel'],
    'Chance_of_winning': [100, 50, 0.01],
    'City': ['Barcelona', 'Manchester', 'Mitzpe Ramon']
}

merge_df1 = pd.DataFrame(data)
```

```python
data = {
    'Team': ['Spain', 'England', 'Israel', 'Slovenia'],
    'num_home_games': [10, 11, 16, 12],
    'avg_players_salary': [100, 200, 300, 400]
}

merge_df2 = pd.DataFrame(data)
```

```python
merge_df1.head()
```

```python
merge_df2.head()
```

## Concats

```python
pd.concat([merge_df1, merge_df2])
```

## Merges

```python
merge_df1.head()
```

```python
merge_df2.head()
```

```python
merge_df1.merge(merge_df2, on='Team')
```

```python
merge_df1.merge(merge_df2, how='outer', on='Team')
```

```python
merge_df1
```

```python
merge_df1.merge(merge_df2, how='left', on='Team')
```

```python
merge_df1.merge(merge_df2, how='right', on='Team')
```
