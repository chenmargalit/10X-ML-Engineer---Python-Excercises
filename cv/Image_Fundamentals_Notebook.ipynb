# Image Fundamentals
This notebook covers the fundamentals of images, including:
- Pixels
- Resolution
- RGB Color Model
- Common Image Formats (JPEG, PNG, etc.)


## 1. Pixels
An image is composed of small units called **pixels**. Each pixel represents a single point of color in the image. 
- **Grayscale Images**: Each pixel has a single intensity value (e.g., 0 for black, 255 for white).
- **Color Images**: Each pixel has three values corresponding to Red, Green, and Blue (RGB).


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grayscale_image = np.array([
    [0, 50, 100, 150, 200],
    [25, 75, 125, 175, 225],
    [50, 100, 150, 200, 250],
    [75, 125, 175, 225, 255],
    [100, 150, 200, 250, 255]
], dtype=np.uint8)
```

```python
plt.imshow(grayscale_image, cmap='gray', vmin=0, vmax=255)
plt.title('Grayscale Image')
plt.colorbar(label='Pixel Intensity')
plt.show()
```

## 2. Resolution
Resolution refers to the total number of pixels in an image, typically described as width × height (e.g., 1920×1080).
- **Higher Resolution**: More detail, larger file size.
- **Lower Resolution**: Less detail, smaller file size.


## 3. RGB Color Model
RGB stands for **Red, Green, Blue**. 
- Each pixel in an RGB image contains three values, one for each channel.
- Colors are created by combining these channels at varying intensities (0 to 255).


```python
rgb_image = np.zeros((5, 5, 3), dtype=np.uint8)
rgb_image[:, :, 0] = 200  # Red channel
rgb_image[:, :, 1] = 200  # Green channel
rgb_image[:, :, 2] = 250   # Blue channel

plt.imshow(rgb_image)
plt.title('RGB Image')
plt.show()
```
